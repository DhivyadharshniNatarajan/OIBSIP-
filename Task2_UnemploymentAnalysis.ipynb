 **Import the necessary libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

**Load the dataset**

```python

df = pd.read_csv("/content/Unemployment in India.csv")
```

**Data Cleaning and Preprocessing**

```python
# Drop any unnecessary columns
df = df.drop(['Area', ' Frequency'], axis=1)
```

```python
# Set ' Date' column as the index
df.set_index(' Date', inplace=True)
```

```python
df[' Estimated Unemployment Rate (%)']
```

```python
# Convert 'Estimated Unemployment Rate (%)' column to numeric type
df[' Estimated Unemployment Rate (%)'] = pd.to_numeric(df[' Estimated Unemployment Rate (%)'], errors='coerce')
```

```python
# Drop rows with missing values
df = df.dropna(subset=[' Estimated Unemployment Rate (%)'])
```

**Analyzing the Unemployment Rate**

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index , df[' Estimated Unemployment Rate (%)'], marker='o')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate in India')
plt.grid(True)
plt.show()
```

**Interactive Plots using Plotly**

```python
# Create a line plot using Plotly
fig = go.Figure()
fig.add_trace(go.Scatter(x=df.index , y=df[' Estimated Unemployment Rate (%)'], mode='lines+markers'))
```

```python
# Customize the plot layout
fig.update_layout(
    title='Unemployment Rate in India',
    xaxis_title='Date',
    yaxis_title='Unemployment Rate (%)'
)
# Show the plot
fig.show()
```
